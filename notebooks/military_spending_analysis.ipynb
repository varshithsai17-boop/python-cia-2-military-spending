```python
import pandas as pd
df = pd.read_csv("../data/military_spending_sipri_2024.csv")
df
```

```python
import matplotlib.pyplot as plt
```

```python
top10_spending = df.sort_values(by='Spending (US$ bn)', ascending=False).head(10)

top10_spending.plot(kind='bar', x='Country', y='Spending (US$ bn)')
plt.title("Top 10 Countries by Military Spending")
plt.xticks(rotation=75)
plt.show()
```

```python
top10_gdp = df.sort_values(by='% of GDP', ascending=False).head(10)

top10_gdp.plot(kind='barh', x='Country', y='% of GDP')
plt.title("Top 10 Countries by % of GDP Spent on Military")
plt.show()
```

```python
# Create spending tiers to group countries
bins = [0, 10, 25, 50, 100, 1000]
labels = ['< $10B', '$10B-$25B', '$25B-$50B', '$50B-$100B', '> $100B']
df['Spending Tier'] = pd.cut(df['Spending (US$ bn)'], bins=bins, labels=labels)

tier_spending = df.groupby('Spending Tier', observed=True)['Spending (US$ bn)'].sum()
tier_spending = tier_spending.sort_values(ascending=False)

tier_spending.plot(kind='bar')
plt.title("Total Military Spending by Tier")
plt.xticks(rotation=45)
plt.show()
```

```python
tier_spending.plot(kind='line', marker='o')
plt.title("Military Spending Trend by Tier")
plt.xticks(rotation=45)
plt.show()
```

```python
top5_global = df.sort_values(by='% of global spending', ascending=False).head(5)

top5_global.set_index('Country')['% of global spending'].plot(kind='pie', autopct='%1.1f%%')
plt.title("Top 5 Countries by Share of Global Military Spending")
plt.ylabel('')
plt.show()
```

```python
df.plot(kind='scatter', x='% of GDP', y='Spending (US$ bn)')
plt.title("Military Spending vs % of GDP")
plt.show()
```

```python
tier_spending.head(5).plot(kind='area')
plt.title("Spending Area View - Top 5 Tiers")
plt.xticks(rotation=45)
plt.show()
```

```python
tier_spending.head(5).plot(kind='bar',
                              color=['#4E79A7','#F28E2B','#59A14F','#E15759','#76B7B2'])
plt.title("Top 5 Spending Tiers - Custom Colors")
plt.xticks(rotation=45)
plt.show()
```
